```python
import numpy as np
import matplotlib.pyplot as plt
```

# 作図の基本

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, color="red", lw=6, ls="-.")
```

```python
plt.plot(x, y, color="red", lw=6, ls="-.")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=sin(x)")
```

## 複数のグラフと凡例表示

```python
y2 = np.cos(x)
plt.plot(x, y, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=sin(x), y=cos(x)")
plt.grid()
plt.savefig("fig01.png")
```
